```python
#import the  requred libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the data set
df=pd.read_csv("revenue_prediction.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
#exmining missing values in a dataset
df.isnull().sum()
```

```python
# drop relevent columns
df=df.drop(columns=["Id","Name","Franchise","Category","City","No_Of_Item"])
```

```python
df.head()
```

```python
df.shape
```

#create independent feature and dependent variable 

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values  
```

#split the dataset into training set and  Test set

```python
pip install scipy
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train ,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
**Predicting the Test set results**
```

```python
y_pred=regressor.predict(x_test)
```

```python
y_pred
```

```python
plt.scatter(x_train,y_train, color="blue")
plt.plot(x_train,regressor.predict(x_train),color="green")
plt.title("revenue vs orders")
plt.xlabel("orders")
plt.ylabel("revenue")
```
